```python
# Importación de librerías necesarias
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

# **Generación de los datos sintéticos**
# Se generan datos con 4 clusters
X, y = make_blobs(n_samples=500, centers=4, cluster_std=0.60, random_state=0)

# Visualización de los datos generados
plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], s=50, cmap='viridis')
plt.title("Datos Sintéticos Generados para Clustering")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()

# **Escalado de los datos**
# El algoritmo K-Means es sensible a la escala de las características, así que se escalan los datos
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# **Aplicación del algoritmo K-Means**
# Se utiliza K-Means con 4 clusters (n_clusters=4) para segmentar los datos
kmeans = KMeans(n_clusters=4, random_state=0)
kmeans.fit(X_scaled)

# **Predicción de etiquetas de los clusters**
y_kmeans = kmeans.predict(X_scaled)

# **Visualización de los clusters y los centroides**
plt.figure(figsize=(8, 6))
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=y_kmeans, s=50, cmap='viridis')
# Mostrar los centroides
centroids = kmeans.cluster_centers_
plt.scatter(centroids[:, 0], centroids[:, 1], c='red', s=200, alpha=0.5, marker='X')
plt.title("Clusters Generados por K-Means")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()

# **Evaluación con Silhouette Score**
# El Silhouette Score evalúa la calidad del clustering. Un valor cercano a 1 indica buenos clusters.
score = silhouette_score(X_scaled, y_kmeans)
print(f"Silhouette Score: {score}")
```
